```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('species',axis=1),df['species'],test_size=0.25,random_state=42)
```

```python
y_train
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
label.fit(y_train)
```

```python
y_train=pd.DataFrame(label.transform(y_train))
y_test=pd.DataFrame(label.transform(y_test))
```

```python
y_train.value_counts()
```

```python
x_train
```

# Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
param={'penalty':['l1','l2'],'C':[1,2,3,4,5,6,10,20,30,40,50]}
gc=GridSearchCV(model,param,scoring='accuracy',cv=5)
```

```python
gc.fit(x_train,y_train)

```

```python
gc.best_score_
```

```python
gc.best_params_
```

```python
y_pred=gc.predict(x_test)
```

```python
y_test
```

```python
  y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Naive Bayes's

```python
from sklearn.naive_bayes import GaussianNB
gauss=GaussianNB()
```

```python
gauss.fit(x_train,y_train)
```

```python
y_pred=gauss.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
