```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('https://github.com/RohanDubeySF/Training/raw/refs/heads/master/EDA/zomato.csv',encoding='latin-1')
data
```

```python
df=pd.read_excel('https://github.com/RohanDubeySF/Training/raw/refs/heads/master/EDA/Country-Code.xlsx')
df
```

```python
final=pd.merge(data,df,on='Country Code',how='left')
final.head()
```

```python
final.columns
```

```python
final.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
final.dropna()
```

```python
final.columns
```

```python
final.Country.value_counts().values
```

```python
plt.figure(figsize=(8,8))
plt.title('Top 3 Countries')
plt.pie(final.Country.value_counts()[:3],labels=final.Country.value_counts().index[:3],autopct='%1.2f%%')
plt.show()
```

```python
final.columns
```

```python
rate=pd.DataFrame(final[['Aggregate rating', 'Rating color', 'Rating text']].value_counts().sort_index(level='Aggregate rating').reset_index())
rate
```

```python
rate.columns
```

```python
rate['Rating color'].unique()
```

```python
plt.figure(figsize=(8,8))
sns.barplot(data=rate,x='Aggregate rating',y='count',hue='Rating color',palette=['black', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
plt.show()
```
