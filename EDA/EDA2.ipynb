```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://github.com/RohanDubeySF/random/raw/refs/heads/main/train%201.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
##HAndling categorical feature Gender
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1 , '18-25':2 , '26-35':3 , '36-45':4 , '46-50':5 , '51-55':6, '55+':7})
df.head()
```

```python
df.City_Category.value_counts()
```

```python
df=pd.concat([df.drop('City_Category',axis=1),pd.get_dummies(df['City_Category'],drop_first=True).replace({True:1,False:0})],axis=1)
df.head()
```

```python
  df.isnull().sum()
```

### Handling Missing values

```python
df['Product_Category_2'].value_counts()
```

```python
df['Product_Category_2'].mode()[0]
```

```python
## Replace the missing values with mode as the data is discrete numerical value
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
df.isnull().sum()
```

```python

df['Product_Category_3'].value_counts()
```

```python
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
df.isnull().sum().sum()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].map({'2':2, '4+':4, '3':3, '1':1, '0':0})
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(['Product_ID','Purchase'],axis=1),df['Purchase'],test_size=0.3,random_state=69)
```

```python
x_train
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=pd.DataFrame(scaler.fit_transform(x_train),columns=[x_train.columns])
x_test=pd.DataFrame(scaler.fit_transform(x_test),columns=[x_test.columns])
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
Linear=LinearRegression()
from sklearn.metrics import r2_score
```

```python
Linear.fit(x_train,y_train)
```

```python
y_pred=Linear.predict(x_test)
```

```python
acc=r2_score(y_test,y_pred)
acc
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
acc=r2_score(y_test,y_pred)
acc
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
acc=r2_score(y_test,y_pred)
acc
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor()
ada.fit(x_train,y_train)
y_pred=ada.predict(x_test)
acc=r2_score(y_test,y_pred)
acc
```
