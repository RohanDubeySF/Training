# **Performing Encoding on Ordinal Data**

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day26-ordinal-encoding/customer.csv")
data.head()
```

```python
data=data.drop(columns=['age','gender'])
data.head()
```

```python
data['review'].unique(), data['education'].unique()
```

# Split Data into test and train and only fit the encoders on Train data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data[['education','review']], data['purchased'],test_size=0.3, random_state=0)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

# Perform Ordinal Encoding on Dependent Columns

```python
from sklearn.preprocessing import OrdinalEncoder
Ord=OrdinalEncoder(categories=[['School', 'UG', 'PG'], ['Poor','Average', 'Good']])
```

```python
Ord.fit(x_train)
```

```python
x_train.head(10)
```

```python
x_train=Ord.transform(x_train)
```

```python
x_train=pd.DataFrame(x_train, columns=['education','review'])
x_train.head(10)
```

```python
Ord.categories_
```

# Perform Label Encoder only on Target Columns

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
label.fit(y_train)
```

```python
y_train.head(10)
```

```python
y_train=pd.DataFrame(label.transform(y_train),columns=['purchased'])
y_train.head(10)
```
