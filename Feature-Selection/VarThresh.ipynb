# Variance Threshold


```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
data=pd.read_csv('/content/sample_data/Santander Customer Satisfaction - TRAIN.csv')
data
```

```python
x=data.drop('TARGET',axis=1)
y=data['TARGET']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=69)
```

```python
from sklearn.feature_selection import VarianceThreshold
var=VarianceThreshold(threshold=0.2)
var.fit(x_train)
```

```python
var.get_support()
```

#### storing all the column names where var > threshold

```python
cols=x_train.columns[var.get_support()]
cols
```

```python
new_x_train=var.transform(x_test)
new_x_train
```

```python
new_x_train = pd.DataFrame(new_x_train, columns=cols)
new_x_train
```
